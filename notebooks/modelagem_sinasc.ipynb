```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

pasta = Path("../dados/processados/resultados_modelagem")
```

```python
metricas = pd.read_csv(pasta / "metricas_modelos.csv")
metricas
```

```python
matriz_log = pd.read_csv(pasta / "matriz_confusao_regressao_logistica.csv", index_col=0)
matriz_flo = pd.read_csv(pasta / "matriz_confusao_floresta_aleatoria.csv", index_col=0)

fig, eixos = plt.subplots(1, 2, figsize=(10, 4))
for eixo, titulo, matriz in zip(eixos, ["Regressao logistica", "Floresta aleatoria"], [matriz_log, matriz_flo]):
    imagem = eixo.imshow(matriz, cmap="Blues")
    eixo.set_title(titulo)
    eixo.set_xticks(range(len(matriz.columns)), matriz.columns, rotation=30, ha="right")
    eixo.set_yticks(range(len(matriz.index)), matriz.index)
    for i in range(matriz.shape[0]):
        for j in range(matriz.shape[1]):
            eixo.text(j, i, int(matriz.iloc[i, j]), ha="center", va="center")
fig.colorbar(imagem, ax=eixos, shrink=0.8)
plt.tight_layout()
```
